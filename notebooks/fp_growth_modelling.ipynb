# 4) FP-Growth modelling

Chạy FP-Growth để khai phá frequent itemsets và association rules.

> Các biến tham số (để papermill inject):
- `DATA_PATH`: đường dẫn file CSV dữ liệu thô.
- `COUNTRY`: lọc theo quốc gia (tuỳ chọn, để giảm kích thước).
- `MIN_SUPPORT`, `MIN_CONFIDENCE`, `MIN_LIFT`: ngưỡng mining.


```python
# Parameters (papermill)
DATA_PATH = "data/online_retail.csv"
COUNTRY = None
MIN_SUPPORT = 0.01
MIN_CONFIDENCE = 0.5
MIN_LIFT = 1.0
```

```python
import pandas as pd
from src.apriori_library import FPGrowthMiner, DataVisualizer

basket_bool = pd.read_parquet("data/processed/basket_bool.parquet").astype(bool)

miner = FPGrowthMiner()
fi, rules, stats = miner.run(
    basket_bool,
    min_support=MIN_SUPPORT,
    metric="confidence",
    min_threshold=MIN_CONFIDENCE
)

print(stats)
rules = rules[rules["lift"] >= MIN_LIFT].sort_values(["lift","confidence"], ascending=False)
rules.head(10)
```

## Trực quan hoá (ví dụ)

```python
viz = DataVisualizer()
viz.plot_top_rules_bar(rules, metric="lift", top_n=10)
viz.plot_support_confidence_scatter(rules)
```

## Lưu kết quả

```python
fi.to_parquet("data/processed/frequent_itemsets_fpgrowth.parquet", index=False)
rules.to_parquet("data/processed/rules_fpgrowth.parquet", index=False)
```
