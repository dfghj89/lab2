# FP-Growth modelling

```
basket_path="data/processed/basket_bool.parquet"
min_support=0.01
min_confidence=0.3
min_lift=1.0
output_rules_path="data/processed/fpgrowth_rules.parquet"
```

```
import time
import pandas as pd
from src.apriori_library import FPGrowthMiner, DataVisualizer
```

```
basket_bool = pd.read_parquet(basket_path)
t0=time.perf_counter(); fi, rules = FPGrowthMiner().run(basket_bool,min_support,min_confidence,min_lift); t1=time.perf_counter()
print("itemsets",len(fi),"rules",len(rules),"time",t1-t0)
rules.head()
```

```
rules.to_parquet(output_rules_path, index=False)
v=DataVisualizer(); v.scatter_support_conf(rules,"FP-Growth: Support vs Confidence"); v.bar_top_lift(rules,15,"FP-Growth: Top lift rules")
```
